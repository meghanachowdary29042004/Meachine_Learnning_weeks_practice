```python
import numpy as np
```

```python
import pandas as pd
```

```python
data=pd.read_excel("IceCreamData.xlsx")
```

```python
x=data['Temperature'].values
```

```python
y=data['Revenue'].values
```

```python
def calculate_slope(x, y):
    n = len(x)
    sum_X = np.sum(x)
    sum_Y = np.sum(y)
    sum_XY = np.sum(x * y)
    sum_X_squared = np.sum(x ** 2)
    
    numerator = (n * sum_XY) - (sum_X * sum_Y)
    denominator = (n * sum_X_squared) - (sum_X ** 2)
    
    slope = numerator / denominator
    return slope
```

```python
def calculate_intercept(x, y, slope):
    n = len(x)
    sum_Y = np.sum(y)
    sum_X = np.sum(x)
    
    intercept = (sum_Y - slope * sum_X) / n
    return intercept
```

```python

slope = calculate_slope(x, y)
intercept = calculate_intercept(x, y, slope)

print(f"Manually Calculated Slope: {slope}")
print(f"Manually Calculated Intercept: {intercept}")
```

```python
def predict(x, slope, intercept):
    return slope * x + intercept

predictions = predict(x, slope, intercept)

# Visualize the data and regression line
import matplotlib.pyplot as plt

plt.scatter(x, y, color='blue')
plt.plot(x, predictions, color='red')
plt.title('Temperature vs Ice Cream Revenue')
plt.xlabel('Temperature')
plt.ylabel('revenue')
plt.show()
```
